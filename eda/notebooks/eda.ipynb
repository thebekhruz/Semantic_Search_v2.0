```python
# Rename columns for better interpretation
df = df.rename(columns={'IAID': 'Image_ID', 'text': 'Document', 'mentions': 'Named_Entities'})
df.head()
```

```python
import re
from typing import List, Dict, Any

def expand_abbreviations(text: str, abbreviation_map: Dict[str, str]) -> str:
    """Expands abbreviations in a given text using the provided mapping."""
    for abbr, full_form in abbreviation_map.items():
        text = re.sub(rf'\b{re.escape(abbr)}\b', full_form, text)
    return text

def process_dataframe(df: pd.DataFrame, abbreviation_map: Dict[str, str]) -> pd.DataFrame:
    """
    Processes the entire dataframe to expand abbreviations in the 'Document' column
    and update the 'Named_Entities' column accordingly.
    
    Args:
        df (pd.DataFrame): The input dataframe.
        abbreviation_map (dict): A dictionary mapping abbreviations to their expanded forms.
    
    Returns:
        pd.DataFrame: The updated dataframe with expanded text and named entities.
    """
    updated_rows = []
    
    for _, row in df.iterrows():
        document = row['Document']
        named_entities = row['Named_Entities']
        
        # Expand abbreviations in document
        updated_document = expand_abbreviations(document, abbreviation_map)
        
        updated_named_entities = []
        for entity in named_entities:
            expanded_span = expand_abbreviations(entity['ne_span'], abbreviation_map)
            
            # If the entity is a date, retain only numeric data
            if entity.get('ne_type') == 'DATE':
                numeric_span = re.sub(r'\D', '', expanded_span)
                updated_named_entities.append({
                    **entity,
                    'ne_span': numeric_span
                })
            else:
                updated_named_entities.append({
                    **entity,
                    'ne_span': expanded_span
                })
        
        updated_rows.append({
            'Image_ID': row['Image_ID'],
            'Document': updated_document,
            'Named_Entities': updated_named_entities
        })
    
    return pd.DataFrame(updated_rows)

# Example abbreviation map
abbreviation_map = {
    # Street Names
    "Rd": "Road ",
    "Rd.": "Road ",
    "St": "Street ",
    "St.": "Street ",
    "Ave": "Avenue ",
    "Ave.": "Avenue ",
    # Other
    'R': 'River ',
    'R.': 'River ',
    # Compass Directions
    "N": "North",
    "S": "South",
    "E": "East",
    "W": "West",
    "SE": "South East",
    "SW": "South West",
    "NE": "North East",
    "NW": "North West",
}


updated_df = process_dataframe(df, abbreviation_map)
updated_df.head()
```

```python
# Save the data
processed_data_filepath = '../../data/processed/processed_data.csv'
updated_df.to_csv(processed_data_filepath, index=False)
print(f"Processed data saved to {processed_data_filepath}")
```
