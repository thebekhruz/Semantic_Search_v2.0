### Exploding Mentions column and looking at items individually. 

```python
# Explode the mentions column
df_exploded = df.explode('mentions')

# Normalize the exploded mentions column
df_exploded_normalized = pd.json_normalize(df_exploded['mentions'])

# Merge the normalized data back with the original DataFrame
df_exploded = df_exploded.drop(columns=['mentions']).reset_index(drop=True)
df_exploded = df_exploded.join(df_exploded_normalized)

del df_exploded_normalized
del df 


df_exploded.head(5)
```

## Normalising Dates

```python
# Output all objects with ny_type: 'DATE' which are not in a YYYY format
df_exploded_date = df_exploded[df_exploded['ne_type'] == 'DATE']
df_exploded_date = df_exploded_date[~df_exploded_date['ne_span'].str.match(r'^\d{4}$')]
df_exploded_date.head()
```

```python
# Now we need normalise the original text. We will use ne_start and ne_end values to extract the original date values from the text column.
# We will replace the original date values with the new YYYY format values.

# Set pandas to display the full text in the 'text' column
print(f"This is an example of text BEFORE normalising the dates:\n {df_exploded_date['text'].iloc[0]}\n")

# Exctract dates in YYYY format from the ne_span column
df_exploded_date['exctracted_date'] = df_exploded_date['ne_span'].str.replace(r'\D', '', regex=True)

# Extract the original date values from the text column using the ne_start and ne_end values
df_exploded_date['original_date'] = df_exploded_date.apply(lambda x: x['text'][x['ne_start']:x['ne_end']], axis=1)

# Replace the original date values with the new YYYY format values
df_exploded_date['text'] = df_exploded_date.apply(lambda x: x['text'].replace(x['original_date'], x['exctracted_date']), axis=1)

# Drop the original date column as we will no longer need it
df_exploded_date = df_exploded_date.drop(columns=['original_date'])

# Read the whole text in the first object
print(f"This is an example of text AFTER normalising the dates:\n {df_exploded_date['text'].iloc[0]}")
df_exploded_date.head()
```

```python
# Save the dataframe to a JSONL file
# We will use cleaned_date_data.jsonl for building clear graph database. 
df_exploded_date.to_json('../../data/processed/cleaned_date_data.jsonl', orient='records', lines=True)
del df_exploded_date
```

## Expanding Abbrivation

```python
df_exploded_date.head()
```

```python
# Load spacy model
import spacy
nlp = spacy.load("en_core_web_sm")

df.head()
```

```python
ABBREVIATION_EXPANSIONS = {
    # Compass directions
    "N": "North",
    "S": "South",
    "E": "East",
    "W": "West",
    "NE": "Northeast",
    "NW": "Northwest",
    "SE": "Southeast",
    "SW": "Southwest",
    # Street Names
    "St": "Street",
    "Rd": "Road",
    "Ave": "Avenue",
    "Dr": "Drive",
    "Blvd": "Boulevard",
    "Ln": "Lane",
    "Ct": "Court",
    "Pl": "Place",
    "Sq": "Square",
    "Ter": "Terrace",
    "Cir": "Circle",
    # Geographical locations (examples)
    "US": "United States",
    "UK": "United Kingdom",
    "CA": "California",
    "NY": "New York",
    # Dates
    "c.": "circa", "c" : "circa",
    "Mon.": "Monday ", "Mon": "Monday",
    "Tue.": "Tuesday ", "Tue": "Tuesday",
    "Wed.": "Wednesday ", "Wed": "Wednesday",
    "Thu.": "Thursday ", "Thu": "Thursday",
    "Fri.": "Friday ", "Fri": "Friday",
    "Sat.": "Saturday ", "Sat": "Saturday",
    "Sun.": "Sunday ", "Sun": "Sunday",
    "Jan.": "January ", "Jan": "January ",
    "Feb.": "February ", "Feb": "February",
    "Mar.": "March ", "Mar": "March",
    "Apr.": "April ", "Apr ": "April",
    "May": "May",
    "Jun.": "June ", "Jun": "June",
    "Jul.": "July ", "Jul": "July",
    "Aug.": "August ", "Aug": "August",
    "Sep.": "September ", "Sep": "September ", "Sept.": "September ", "Sept": "September ",
    "Oct.": "October ", "Oct": "October",
    "Nov.": "November ", "Nov": "November",
    "Dec.": "December ", "Dec": "December"
    # Extend this list as needed
}


def expand_abbreviations(text):
    for abbr, expanded in ABBREVIATION_EXPANSIONS.items():
        text = text.replace(abbr, expanded)
    return text

# Expand abbreviations in the text column
df['text'] = df['text'].apply(expand_abbreviations)

# Remove any extra whitespace
df['text'] = df['text'].str.strip()
```

```python
# Lowercase all text in the text column
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
# Remove all punctuatuation from the text column using spacy
df['text'] = df['text'].apply(lambda x: " ".join([token.text for token in nlp(x) if not token.is_punct]))
```

```python
# Remove all stop words from the text column using spacy
df['text'] = df['text'].apply(lambda x: " ".join([token.text for token in nlp(x) if not token.is_stop]))
```

```python
# Remove any urls from the text column
df['text'] = df['text'].str.replace(r'http\S+', '', regex=True)
```

```python
# Remove any extra whitespace
df['text'] = df['text'].str.strip()
```
